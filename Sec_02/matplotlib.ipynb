```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20)
y = x ** 2
z = x + y
```

```python
plt.plot(x, y, x, x)
```

```python
plt.subplot(3, 2, 1)
plt.plot(x, x)
plt.subplot(3, 2, 2)
plt.plot(x, y)
plt.subplot(3, 2, 3)
plt.plot(x, z)
plt.subplot(3, 2, 4)
plt.plot(y, z)
plt.subplot(3, 2, 5)
plt.subplot(3, 2, 6)
plt.tight_layout()
```

```python
plt.scatter(x, y)
```

```python
plt.hist(x, y)
```

```python
plt.bar(x, y)
```

```python
plt.fill(x, y)
```

```python
plt.polar(x, y)
```

```python
plt.plot(x, y, y, x, label='l1')
plt.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .8, .6])
plt.plot(x, y, y, x)
```

```python
fig, axes = plt.subplots(2, 2)

axes[(0, 0)].plot(x, y)
axes[(1, 1)].plot(x, y, y, x)

plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .8, .6])
axes.set_xlabel('truc x')
axes.set_ylabel('truc y')
axes.set_title('minh hoa')
plt.plot(x, y, y, x)
```
